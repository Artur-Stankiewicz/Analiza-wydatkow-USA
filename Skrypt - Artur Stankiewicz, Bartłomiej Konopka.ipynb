### Następnie na podstawie tych danych sporządziliśmy średnią dla stanów o wysokim i niskim poziomie RDO oraz policzyliśmy różnicę pomiędzy tymi dwoma średnimi. Na podstawie tych danych doszliśmy do następujących wniosków:
- różnica między stanami o wysokim rozporządzalnym dochodem osobistym (RDO) a niskim RDO była na przestrzeni lat na mniej więcej tym samym poziomie

### W kolejnej części postanowiliśmy przyjrzeć się strukturze wydatków w poszczególnych stanach w 2018 roku na podstawowe potrzeby takie jak:
- żywność
- opieka zdrowotna
- zamieszkanie
- paliwo i inne odbra energetyczne
### Na podstawie tych danych doszliśmy do wniosków, że:
- koszty żywności oraz paliwa i innych dóbr energetycznych utrzymują się na pododbnym poziomie
- ale koszty zamieszkania oraz opieki zdrowotnej znacząco odbiegają od siebie
- koszty opieki zdrowotnej są generalnie wysokie, niezależnie od analizowanego stanu, co może stanowić wykluczenie z dostępu do opieki zdrowotnej osób o najniższym dochodzie
- niewątpliwie koszty związane z zamieszkaniem oraz koszty opieki zdrowotnej kształtują realny poziom życia w poszczególnych stanach

Na podstawie danych z BEA dotyczących rozkładu dochodu w społeczeństwie w 2018 roku dla całych Stanów Zjednoczonych stworzyliśmy wykres przedstawiający Krzywą Lorenza. # _Krzywa Lorenza w USA w 2018 roku_

### W tej części:
- Wczytujemy bazę danych
- Sprawdzamy nazwy kolumn
- Sprawdzamy wymiar bazy danych
- Sprawdzamy potencjalne missingi
- Sprawdzamy w jakiej postaci mamy pogrupowany rozkład dochodu w społeczeństwie

```python
Lorenz = pd.read_csv(r"C:\Users\Artur\Desktop\Python pliki\Projekt\Rozkład dochodów w USA.csv")
```

```python
Lorenz.columns
```

```python
Lorenz.shape
```

```python
if Lorenz.isna().sum().sum() == 0:
    print('Baza danych nie zawiera missingów')
else:
    print('Baza danych ma braki')
```

```python
NazwyGrup = Lorenz['Grupa'].unique()
```

```python
NazwyGrup
```

### W tej części:
- Tworzymy kolumnę "Grupy - skumulowane', aby przedstawić kwantyle"
- Wykorzystujemy funkcję cumsum(), aby na podstawie poniższych danych stworzyć kwantyle
- Tworzymy wykres przedstawiający krzywą Lorenza. W tym celu wykorzystujemy dane z tabeli oraz prostą y=x

```python
Lorenz['Grupy - skumulowane'] = np.linspace(0, 1, 11)
```

```python
Lorenz['Kwantyle'] = Lorenz['Procentowy udzial'].cumsum()
```

```python
Lorenz
```

```python
plt.figure(figsize=(15, 12))
sns.lineplot(x='Grupy - skumulowane', y ='Kwantyle', data=Lorenz, marker='s', label = 'Krzywa Lorenza', color = 'darkmagenta')
plt.plot([0, 1], [0, 1], color='lightseagreen', linewidth=2, label="Równomierny rozkład dochodów w społeczeństwie")

plt.title('Krzywa Lorenza dla USA',
          loc='left', fontsize = 20, fontweight='bold', pad = 15, fontname="Times New Roman")

plt.xticks(np.linspace(0,1,11))
plt.yticks(np.linspace(0,1,11))

plt.xlabel('Kwantyle', fontsize = 15, fontname="Times New Roman")
plt.ylabel('')
plt.ylim(ymin=0)
plt.xlim(xmin=0)
plt.ylim(ymax=1)
plt.xlim(xmax=1)
plt.legend(fontsize='large')
plt.grid(ls='--')
plt.savefig('Lorenz.png')

plt.show()
```
